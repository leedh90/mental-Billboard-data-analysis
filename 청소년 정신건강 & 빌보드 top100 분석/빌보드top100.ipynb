```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../Downloads/charts.csv") # data 불러오기
df.set_index("date", inplace=True) # 인덱싱 변경(날짜기준) 번호로 되어있던 인덱싱을 날짜로 바꿔준다.
df
```

```python
df1 = df.loc["2020-04-03":] # 날짜로 슬라이싱 2020-04-03 이후의 data 값만 추출
df1 # 해당 data만 사용하기 위해 변수저장
```

```python
df1["rank"] == 1 # rank가 1인 값만 추출(boolean 타입으로 추출)
df_rank1 = df1[df1["rank"] == 1].copy() #True 값인 것만 df1에서 추출해서 df_rank1에 넣어줘라.
df_rank1
```
